# 2. Neural Network with Sequential and Functional API

```python
import os
os.environ["TFF_CPP_MIN_LOG_LEVEL"] = "2"
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
print(x_train[0].shape)
```

```python
x_train = x_train.reshape(-1, 784)
print(x_train.shape)
print(x_train.dtype)
x_train = x_train.astype("float32") / 255.0
print(x_train.dtype)
x_test = x_test.reshape(-1, 784)
x_test = x_test.astype("float32") / 255.0
```

```python
# Nếu ở dạng numpy array thì việc chuyển qua tensor là không cần thiết.
x_train = tf.convert_to_tensor(x_train)
print(type(x_train))
```

## Sequential API
Very convenient, not very flexible.

```python
a = tf.random.normal((10, ))
b = tf.random.normal((10, 1))
c = tf.random.normal((1, 10))
print(a)
print(b)
print(c)
```

```python
model = keras.Sequential([
    keras.Input(shape=(784, )),
    layers.Dense(units=512, activation="relu"),
    layers.Dense(units=256, activation="relu"),
    layers.Dense(10),
])

print(model.summary())
```

```python
model = keras.Sequential()
model.add(layers.Input(shape=(784, )))
model.add(layers.Dense(512, "relu"))
model.add(layers.Dense(256, "relu"))
model.add(layers.Dense(10))
print(model.summary())
```

```python
# Compile model: config cấu hình training cho model
model.compile(
    # from_logits=True -> cho output qua softmax trước khi tính loss
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"],
)
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=5, verbose=2)
model.evaluate(x_test, y_test, batch_size=32, verbose=2)
```

## Functional API
A bit more flexible

```python
inputs = layers.Input(shape=(784, ))
x = layers.Dense(512, activation='relu', name="1_dense")(inputs)
x = layers.Dense(256, activation='relu', name='2_dense')(x)
outputs = layers.Dense(10, activation='softmax', name='3_out')(x)

model = keras.Model(inputs=inputs, outputs=outputs)
print(model.summary())
```

```python
# Compile model: config cấu hình training cho model
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"],
)
```

```python
model_debug = keras.Model(inputs = model.inputs,
                          outputs = [layer.output for layer in model.layers])

features = model_debug.predict(x_train)

for feature in features:
    print(feature.shape)
```
