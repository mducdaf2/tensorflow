# 1. Basic of TensorFlow

```python
import os
import tensorflow as tf
```

## Initialization Tensor

```python
x = tf.constant(4, shape=(1, 1), dtype=tf.float32)
print(x)
print(type(x))
print(x.shape)
print(x.dtype)
```

```python
x = tf.constant([[1, 2, 3], [3, 4, 5]])
print(x)
```

`tf.ones()`

```python
x = tf.ones((3, 3))
print(x)
print(type(x))
```

`tf.eye()`

```python
x = tf.eye(4)
print(x)
```

`tf.random.normal()`

```python
x = tf.random.normal((3, 4), mean=0, stddev=1)
print(x)
```

`tf.random.uniform()`

```python
x = tf.random.uniform((1, 3), minval=0, maxval=3)
print(x)
```

`tf.range()`

```python
print(tf.range(1, 10, 2))
```

`tf.cast()`

```python
x = tf.range(10)
print(x)
print(x.dtype)
y = tf.cast(x, dtype=tf.float16)
print(y)
print(y.dtype)
```

## Mathematical Operations

```python
x = tf.constant([1, 2, 3])
y = tf.constant([9, 8, 7])

z = tf.tensordot(x, y, axes=0)
print(z)
```

```python
print(tf.reduce_sum(z, axis=1))
```

`tf.matmul()`

```python
x = tf.random.normal((3, 4))
y = tf.random.normal((4, 5))
print(tf.matmul(x, y))
print(x @ y)
```

## Indexing

```python
x = tf.range(10)
print(x[:])
print(x[2:])
print(x[:4])
print(x[::2])
print(x[::-1])
```

```python
indices = tf.constant([1, 3])
y = tf.gather(x, indices)
print(y)
```

```python
x = tf.constant([
    [2, 3, 4],
    [1, 2, 3]
])

print(x[:, 1:2])
```

## Reshaping

`tf.reshape()`

```python
x = tf.range(9)
print(x)
y = tf.reshape(x, shape=(3, 3))
print(y)
```

`tf.transpose()`

```python
z = tf.transpose(y, perm=[1, 0])
print(z)
```
